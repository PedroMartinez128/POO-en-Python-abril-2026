# Lenguaje Python , Crear la clase "Animal" con los atributos "Nombre", "Raza" crear el contructor animal, y las funciones Ladrar(), Correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El Constructor (__init__) inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (Método) Ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!😍"  )

    # Función (Método) Correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.😎")
    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina🤣")

# --- Creación de los objetos (Instanciación) ---

# Objeto 1: Pitbull
pitbull = Animal("Tyson", "Pitbull")

# Objeto 2: Chihuahua

chihuahua = Animal("Chispa", "Chihuahua")

# Creamos el objeto Perro callegero con datos
perrocallegero = Animal("Solovino", "Chafa")

# --- Uso de los objetos ---

print(f"Datos del primer perro: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()


print("-" * 20)

print(f"Datos del segundo perro: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 20)

print(f"Datos del Tercer perro: {perrocallegero.nombre}, Raza: {perrocallegero.raza}")
perrocallegero.ladrar()
perrocallegero.correr()
perrocallegero.Morder()

print("Programa realizado por Pedro Martinez NL = 34")
```
